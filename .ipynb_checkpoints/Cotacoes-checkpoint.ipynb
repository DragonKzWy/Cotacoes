# Pegando a Cotação e uma ação

```python
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yfin
import matplotlib.pyplot as plt

yfin.pdr_override()

cotacao_bovespa = pdr.get_data_yahoo('^BVSP', start='2020-01-01', end='2021-01-01')

display(cotacao_bovespa)

cotacao_bovespa["Adj Close"].plot(figsize=(12, 8))

plt.show()

cotacao_petrobras = pdr.get_data_yahoo('PETR4.SA', start='2020-01-01', end='2021-01-01')

display(cotacao_petrobras)

cotacao_petrobras["Adj Close"].plot(figsize=(12, 8))

plt.show()
```

```python
tabela_empresas = pd.read_excel("Empresas.xlsx")
display(tabela_empresas)

for empresa in tabela_empresas['Empresas']:
    print(empresa)
    
    cotacao = pdr.get_data_yahoo(f'{empresa}.SA', start='2020-01-01', end='2021-01-01')

    display(cotacao)

    cotacao["Adj Close"].plot(figsize=(12, 8))

    plt.show()
```
